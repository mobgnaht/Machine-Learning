**Câu 1 :**  Vẽ dendrogram và chọn số cụm thích hợp cho 4 distances : \\
Về việc chọn số lượng cluster, ta sẽ chọn cái khoảng cách lớn nhất của 2 cluster, sau đó cắt ngang chúng thì số lượng đường đi ra khỏi cái đường cắt ngang mà nằm bên dưới đường này sẽ là số lượng cluster. 


*   WARD METHOD \\
Đối với trường hợp này , ta chọn n_clusters = 5 ( có thể chọn n_cluster = 3 nhưng trường hợp này số cluster hơi ít).



```python
import scipy.cluster.hierarchy as sch
# make a dendogram used ward method 
plt.figure(figsize=(12,6),dpi = 80)
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))

plt.axhline(y = 170,color = 'r',linestyle = '--')

plt.title('Dendogram - Ward method')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')
plt.show()
```


*   AVERAGE METHOD \\



*   COMPLETE METHOD \\

**Câu 2 :** Vẽ hình biểu diễn các cụm đối với cả 4 distances :

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans

def plot_clustering(X,n_clusters,linkage):
    labels = AgglomerativeClustering(n_clusters=n_clusters,affinity= 'euclidean', linkage = linkage)
    labels.fit(X)
    y_labels = labels.fit_predict(X)
    kmeans = KMeans(n_clusters=n_clusters).fit(X)
    label = np.unique(y_labels)
  
    plt.figure(figsize=(12,6),dpi = 80)
    for i in label:
        plt.scatter(X[y_labels==i,0],X[y_labels == i,1],label = 'cluster_'+str(i+1))

    plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='*',s = 200,c = 'red',label = 'centroid')

    plt.xlabel('Annual Income')
    plt.ylabel('Spending Score')
    plt.title('Hierarchical Clustering')
    plt.legend(loc = 'best')
```



*   WARD METHOD



```python
n_clusters = 5
plot_clustering(X,n_clusters,'ward')
plt.show()
```



*   AVERAGE METHOD



```python
n_clusters = 7
plot_clustering(X,n_clusters,'average')
plt.show()
```



*   COMPLETE METHOD



```python
n_clusters = 4
plot_clustering(X,n_clusters,'complete')
plt.show()
```



*   SINGLE METHOD



```python
n_clusters = 8
plot_clustering(X,n_clusters,'single')
plt.show()
```
