```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/huynhthanh98/ML/master/lab-04/train.csv')
data.head()
```

```python
data=data.drop(["Cabin","Age","PassengerId","Name","Ticket"],axis=1)
data=data.dropna()
```

```python
print(data)
```

```python
data_train, data_test = train_test_split( data, test_size=0.2, random_state=42)
```

```python
def get_features(data=data_train):
  Survived=data["Survived"].values
  Pclass=data["Pclass"].values
  Sex=data["Sex"].values
  SibSp=data["SibSp"].values
  Parch=data["Parch"].values
  Embarked=data["Embarked"].values
  Fare=data["Fare"].values
  return Survived, Pclass, Sex, SibSp, Parch, Embarked,Fare

Survived_train, Pclass_train, Sex_train, SibSp_train, Parch_train,Embarked_train,Fare_train=get_features()
Survived_test, Pclass_test, Sex_test, SibSp_test, Parch_test,Embarked_test,Fare_test=get_features(data=data_test)
```

```python
#Phân tích thử các giá trị có trong mỗi feature
for i in data_train:
  print(i)
  print(data_train[i].unique())
```

**Câu 1 :**


```python
# Xác suất tiên nghiệm 
P_Survived = np.sum((Survived_train == 1))/len(Survived_train)
P_notSurvived = np.sum((Survived_train == 0))/len(Survived_train)
print(data_test)
```

```python
features = ["Sex","Pclass","Embarked"]

P_Survived_combine = np.ones(data_test.shape[0])
P_notSurvived_combine = np.ones(data_test.shape[0])
for i in features:
  col_train = data_train[i]
  col_test = data_test[i]

  P_1 = []
  P_0 = []
  for j in col_train.unique():
    
    P1 = np.sum((col_train[Survived_train ==1]==j))/np.sum((Survived_train == 1))
    P_1.append(P1)
    P0 = np.sum((col_train[Survived_train ==0]==j))/np.sum((Survived_train == 0))
    P_0.append(P0)

  P_col_Survived = np.zeros(col_test.shape)
  P_col_notSurvived = np.zeros(col_test.shape)

  ##Tính xác xuất hậu nghiệm trên tập trên tập test)
  
  for k in range(len(col_test.unique())): 
    P_col_Survived += P_1[k]*(col_test==col_train.unique()[k]).astype(float) 
    P_col_notSurvived += P_0[k]*(col_test==col_train.unique()[k]).astype(float)
  
  ## Tính P(x1,...,xn | Survived) = P(x1|Survived) *...*P(xn | Survived)
  P_Survived_combine *= P_col_Survived
  P_notSurvived_combine *= P_col_notSurvived
  
## Tính P(x1,...,xn | Survived) * P(Survived)
P_Survived_combine =P_Survived_combine * P_Survived
P_notSurvived_combine *= P_notSurvived

## Kết quả
np.mean((P_Survived_combine>P_notSurvived_combine).astype(float)==Survived_test)
```

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
for i in features:
  data_train1[i] = le.fit_transform(data_train1[i])
  data_test1[i] = le.fit_transform(data_test1[i])

print(data_test1)
```

```python
nb = CategoricalNB(alpha=1.0e-10).fit(data_train1, Survived_train)
print(nb.score(data_test1,Survived_test))
```

**Câu 2 :**

```python
def normal(x, mean, std):
    return 1/(np.sqrt(2*np.pi)*std)*np.exp(-np.square(x-mean)/(2*np.square(std)))
```

```python
features = ["SibSp","Parch","Fare"]
P_col_Survived = 1
P_col_notSurvived = 1
for i in features:
  col_train = data_train[i]
  col_test = data_test[i]

  ## Tính Tham Số

  col_mean_Survived = np.mean(col_train[Survived_train == 1])
  col_std_Survived = np.std(col_train[Survived_train == 1])

  col_mean_notSurvived = np.mean(col_train[Survived_train == 0])
  col_std_notSurvived = np.std(col_train[Survived_train == 0])

  P_col_Survived *= normal(col_test,col_mean_Survived,col_std_Survived)
  P_col_notSurvived *= normal(col_test,col_mean_notSurvived,col_std_notSurvived)

P_Survived_combine=P_Survived*P_col_Survived
P_notSurvived_combine= P_notSurvived*P_col_notSurvived

print(np.mean((P_Survived_combine>P_notSurvived_combine).astype(int)==Survived_test))
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dinhvietcuong1996/Lab-MachineLearningCourse/master/Lab04/lienminh.csv")
df.head()
```

```python
killsDiff = df['killsDiff'].values
minionsKilledDiff = df['minionsKilledDiff'].values
wardPlacedDiff = df['wardPlacedDiff'].values
firstBlood = df['firstBlood'].values
heralds = df['heralds'].values
dragons = df['dragons'].values

teamWins = df['teamWins'].values

## P(teamWins)

P_blueWins=0.5
P_redWins=0.5
```

```python
#Phân tích thử các giá trị có trong mỗi feature
for i in df:
  print(i)
  print(df[i].unique())
```

```python
## Tính P(Categorical)
features = ["firstBlood","heralds","dragons"]

P_blueWins_combine1 = np.ones(df.shape[0])
P_redWins_combine1 = np.ones(df.shape[0])
for i in features:
  col_train = df[i]
  col_test = df[i]

  P_1 = []
  P_0 = []
  for j in col_train.unique():
    
    P1 = np.sum((col_train[teamWins =='blue']==j))/np.sum((teamWins == 'blue'))
    P_1.append(P1)
    P0 = np.sum((col_train[teamWins == 'red']==j))/np.sum((teamWins == 'red'))
    P_0.append(P0)

  ##Tính xác xuất hậu nghiệm trên tập trên tập test

  P_col_blueWins = np.zeros(col_test.shape)
  P_col_redWins = np.zeros(col_test.shape)
  for k in range(len(col_test.unique())): 
    P_col_blueWins += P_1[k]*(col_test==col_train.unique()[k]).astype(float) 
    P_col_redWins += P_0[k]*(col_test==col_train.unique()[k]).astype(float)
  
  ## Tính P(x1,...,xn | teamWins) = P(x1|teamWins) *...*P(xn | teamWins)
  P_blueWins_combine1 *= P_col_blueWins
  P_redWins_combine1 *= P_col_redWins
```

```python
## Tính P(Gaussian)
features = ['killsDiff','minionsKilledDiff','wardPlacedDiff']

P_col_blueWins = 1
P_col_redWins = 1
for i in features:
  col_train = df[i]
  col_test = df[i]

  ## Tính Tham Số

  col_mean_blueWins = np.mean(col_train[teamWins == 'blue'])
  col_std_blueWins = np.std(col_train[teamWins == 'blue'])

  col_mean_redWins = np.mean(col_train[teamWins == 'red'])
  col_std_redWins = np.std(col_train[teamWins == 'red'])

  ## Tính P(x1,...,xn | teamWins) = P(x1|teamWins) *...*P(xn | teamWins)
  
  P_col_blueWins *= normal(col_test,col_mean_blueWins,col_std_blueWins)
  P_col_redWins *= normal(col_test,col_mean_redWins,col_std_redWins)
```

```python
## Tính P = P(teamWins)*P(Categorical)*P(Gaussian)

P_blueWins_combine=P_blueWins*P_col_blueWins*P_blueWins_combine1
P_redWins_combine=P_redWins*P_col_redWins*P_redWins_combine1
```

```python
# Team Blue Wins

print(np.mean((P_blueWins_combine>P_redWins_combine).astype(float)==(teamWins == 'blue').astype(float)))
```

```python
# Team Red Wins

print(np.mean((P_redWins_combine>P_blueWins_combine).astype(float)==(teamWins == 'blue').astype(float)))
```
