```python
from sklearn import datasets
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
diabetes_X_train,diabetes_y_train=diabetes_X[:-2],diabetes_y[:-2].reshape(1,-1).T
diabetes_X_test,diabetes_y_test=diabetes_X[-2:],diabetes_y[-2:].reshape(1,-1).T
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

**Câu 1:**

```python
one = np.ones((diabetes_X_train.shape[0], 1))
Xbar = np.concatenate((one, diabetes_X_train), axis = 1)

# Tính W
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, diabetes_y_train)
w = np.dot(np.linalg.pinv(A), b)
print('w = ', w)
```

**Câu 2:**

```python
print(diabetes_X_test.shape)
print(w.shape)
one_2 = np.ones((diabetes_X_test.shape[0], 1))
Xbar_2 = np.concatenate((one_2,diabetes_X_test), axis = 1)

# Tìm target values 
target_values = np.dot(Xbar_2,w)
print('target values =',target_values)
print(diabetes_y_test)

# So sánh với với target values trong diabetes_y_test
dist = np.linalg.norm(target_values-diabetes_y_test)
print('Mean squared error: %.4f' % mean_squared_error(diabetes_y_test, target_values))
print('Coefficient of determination: %.4f' % r2_score(diabetes_y_test, target_values))
```

**Câu 3:**

```python
# Create linear regression object
regr = linear_model.LinearRegression()
 
# Train the model 
regr.fit(diabetes_X_train, diabetes_y_train)
 
# Tìm target values của diabetes_X_test
diabetes_y_pred = regr.predict(diabetes_X_test)
 
# Ma trận W 
print('Coefficients: \n', regr.coef_)
print('Intercept: \n', regr.intercept_)
# The coefficients
print('Mean squared error: %.4f' % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.4f' % r2_score(diabetes_y_test, diabetes_y_pred))
```
